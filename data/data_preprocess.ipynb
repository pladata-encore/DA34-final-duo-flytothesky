```python
import os
import concurrent.futures


folder_path = '/content/drive/My Drive/02.라벨링데이터'

def process_file(file_path):
    # 파일 내용 읽기 및 처리 로직
    with open(file_path, 'r') as f:
        content = f.read()
    return content

def batch_process(file_paths):
    results = []
    with concurrent.futures.ThreadPoolExecutor() as executor:
        futures = [executor.submit(process_file, fp) for fp in file_paths]
        for future in concurrent.futures.as_completed(futures):
            results.append(future.result())
    return results

# 폴더 내 파일 목록 가져오기
files = os.listdir(folder_path)

# 필요한 파일 경로 목록 생성 (예: 텍스트 파일만)
file_paths = [os.path.join(folder_path, f) for f in files if f.endswith('.json')]

# 배치 크기 설정 (한 번에 처리할 파일 수)
batch_size = 100

# 배치 처리
for i in range(0, round(len(file_paths)/1000), batch_size):
    batch = file_paths[i:i + batch_size]
    results = batch_process(batch)
    # 배치 처리 결과 사용
    for result in results:
        print(result)

```
